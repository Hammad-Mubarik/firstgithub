# Show Labels

```python
tip = sns.load_dataset("tips")

sns.boxplot(x="day",y="tip", data=tip,hue="smoker",
            showmeans=True,
            meanprops={"marker":"*","markersize":"10","markeredgecolor":"blue"}) 

#show labels


plt.xlabel("Which Day")
plt.ylabel("how much Tip")
plt.title("Boxplot for Tips in Hotel")
```

# Sizes of Labels and Title

```python
tip = sns.load_dataset("tips")

sns.boxplot(x="day",y="tip", data=tip,hue="smoker",
            showmeans=True,
            meanprops={"marker":"*","markersize":"10","markeredgecolor":"blue"}) 

#show labels


plt.xlabel("Which Day",size=12)
plt.ylabel("how much Tip",size=12)
plt.title("Boxplot for Tips in Hotel",size=17)
```

# Making Bold Letters


```python
tip = sns.load_dataset("tips")

sns.boxplot(x="day",y="tip", data=tip,hue="smoker",
            showmeans=True,
            meanprops={"marker":"*","markersize":"10","markeredgecolor":"blue"}) 

#show labels


plt.xlabel("Which Day",size=12,weight="bold")
plt.ylabel("how much Tip",size=12,weight="bold")
plt.title("Boxplot for Tips in Hotel",size=17,weight="bold")
```
