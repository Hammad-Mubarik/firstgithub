## Adding Titles

```python
# import libraries

import seaborn as sns
import matplotlib.pyplot as plt

# load dataset

phool = sns.load_dataset("iris")

# draw a line plot

p= sns.lineplot(x="petal_length",y="petal_width",data=phool)

p.set_title("Phoolon ka plot")               ##add title to plt will be better
plt.show()
```

## Adding x and y Limits

```python
# import libraries

import seaborn as sns
import matplotlib.pyplot as plt

# load dataset

phool = sns.load_dataset("iris")

# draw a line plot

sns.lineplot(x="petal_length",y="petal_width",data=phool)
plt.title("phoolon ka plot")
plt.xlim(3)
plt.ylim(1)
plt.show()
```

## Set Styles

- darkgrid
- whitegrid
- dark
- white
- ticks

```python
# import libraries

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style(style=None, rc=None)

# load dataset

phool = sns.load_dataset("iris")

# draw a line plot

sns.lineplot(x="petal_length",y="petal_width",data=phool)
plt.title("phoolon ka plot")
sns.set_style("whitegrid")
plt.show()
```

## Size of Figure

```python
# import libraries

import seaborn as sns
import matplotlib.pyplot as plt

# load dataset

phool = sns.load_dataset("iris")

# Change figure size

plt.figure(figsize=(10,8))

# draw a line plot

sns.lineplot(x="petal_length",y="petal_width",data=phool)
plt.title("phoolon ka plot")

plt.show()
```
