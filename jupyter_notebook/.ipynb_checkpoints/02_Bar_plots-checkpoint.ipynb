## Numpy Import mean and median

```python
# # import libraries

import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt

# load dataset

titanic = sns.load_dataset("titanic")
titanic

# draw a bar plot

sns.barplot(x="class",y="fare",data=titanic, hue="sex",color="blue",palette = "dark",
           estimator = mean)

plt.title("Titanic ka plot")

plt.show()
```

```python
# # import libraries

import seaborn as sns
from numpy import median
import matplotlib.pyplot as plt

# load dataset

titanic = sns.load_dataset("titanic")
titanic

# draw a bar plot

sns.barplot(x="class",y="fare",data=titanic, hue="sex",color="blue",palette = "dark",
           estimator = median)

plt.title("Titanic ka plot")

plt.show()
```

## Stauration and Estimator = only numeric values not ""

```python
# # import libraries

import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt

# load dataset

titanic = sns.load_dataset("titanic")
titanic

# draw a bar plot

sns.barplot(x="class",y="fare",data=titanic, hue="sex",color="blue",palette = "dark",saturation=.4)
    

plt.title("Titanic ka plot")

plt.show()
```

# Different face line widht and color

## Facecolor = RGBA   ( A= alpha)

```python
# # import the required  libraries

import seaborn as sns
import matplotlib.pyplot as plt


# load dataset

titanic = sns.load_dataset("titanic")
titanic

# draw a bar plot

sns.barplot(x="class",y="fare",data = titanic,linewidth=4,facecolor=(0.4,0.2,1,0.4),errcolor=".3",edgecolor="0")
           

plt.title("Titanic ka plot")

plt.show()
```
